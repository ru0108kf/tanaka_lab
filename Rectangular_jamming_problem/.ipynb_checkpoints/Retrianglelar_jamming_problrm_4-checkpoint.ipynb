```python
#モジュールの読み込み
import matplotlib.pyplot as plt
from matplotlib import patches
import math
import numpy as np
import random
import pandas as pd
import itertools
from IPython.display import clear_output
import time
import array
import numpy
%matplotlib inline
```

```python
#seed値を設定（ここをなくせば完全ランダムになる。)
seed = 0
random.seed(seed)
np.random.seed(seed)
```

```python
#数値(W:部屋の幅,B:長方形の数,V:長方形の面積の合計)
W = 4
B = 20
V = 20
wh = np.random.uniform(low=0.5, high=1.5, size=(B, 2))
wh = wh*(math.sqrt(V/np.sum(wh[:,0]*wh[:,1])))
xy = [[0,0]]*B
```

```python
#DataFrame処理
wh_df = pd.DataFrame(data=(wh),columns=("w","h"))
xy_df = pd.DataFrame(data=(xy),columns=("x","y"))
S_df = wh_df["w"]*wh_df["h"]
Rect = pd.concat([wh_df,xy_df],axis=1)
Rect["x+w"] = Rect["x"]+Rect["w"]
Rect["y+h"] = Rect["y"]+Rect["h"]
Rect
```

```python
#Axesオブジェクトに長方形を配置
def Rect_place(rect,n):
    global ax,fig
    ax.grid()
    colors="#"+''.join([random.choice('0123456789ABCDEF') for i in range(3)])
    r = patches.Rectangle( xy=(rect.loc[n]["x"],rect.loc[n]["y"]) , width=rect.loc[n]["w"], height=rect.loc[n]["h"], color=colors)
    ax.add_patch(r)
```

```python
#index：ｎがそれより若いindexの長方形と重なっていないか確認.重なっていなければ、Trueを返す
def overlap(rect,n):
    result = []
    for i in range(n):
        xn = rect.loc[n]["x"]
        xi = rect.loc[i]["x"]
        wn = rect.loc[n]["w"]
        wi = rect.loc[i]["w"]
        yn = rect.loc[n]["y"]
        yi = rect.loc[i]["y"]
        hn = rect.loc[n]["h"]
        hi = rect.loc[i]["h"]
        result.append(not((max(xn,xi)<min(xn+wn,xi+wi))and(max(yn,yi)<min(yn+hn,yi+hi))))#重なっていなければ、True
    return all(result)#一つでもFalseがあればFalse
```

```python
#index:nが範囲内にあるか確認.範囲内にあれば、Trueを返す
def Out(rect,n):
    xn = rect.loc[n]["x"]
    yn = rect.loc[n]["y"]
    wn = rect.loc[n]["w"]
    hn = rect.loc[n]["h"]
    return xn>=0 and yn>=0 and xn+wn<=W
```

```python
#配置する座標を決める
def place_coordinate(rect,n):
    #x列とx+w列を縦につなげ、効率化のため、重複は消す。
    x_coordinate = pd.concat([rect.loc[:,"x+w"], rect.loc[:,"x"]],axis=0).drop_duplicates()
    y_coordinate = pd.concat([rect.loc[:,"y"], rect.loc[:,"y+h"]],axis=0).drop_duplicates()
    #xとyのすべての組み合わせを列挙する。
    coordinate = pd.DataFrame(itertools.product(x_coordinate, y_coordinate)).rename(columns={0:'x',1:'y'})
    #先にyについてソートした後に、xについてソートする。
    coordinate_sort = coordinate.sort_values(by=["y","x"], ascending=True)
    #行を再ラベリングする。
    coordinate_sort = coordinate_sort.set_axis(list(np.arange(0, len(coordinate))), axis=0)
    i = 0
    while True:
        rect.at[n,"x"] = coordinate_sort.loc[i]["x"]
        rect.at[n,"y"] = coordinate_sort.loc[i]["y"]
        if overlap(rect,n) and Out(rect,n):#重ならず、範囲内にあれば、True
            break
        if i == len(coordinate):#n回繰り返したらやめる
            break
        i += 1
    #x+w列を更新する
    rect["x+w"] = rect["x"]+rect["w"]
    rect["y+h"] = rect["y"]+rect["h"]
```

```python
#実行

# 処理前の時刻
t1 = time.time() 

Rect2 = Rect.copy()
for j in range(B):
    place_coordinate(Rect2,j)
#高さを定義
Height = round(Rect2.max()["y+h"],2)
H = math.ceil(Height)
#Axesオブジェクトを作成
fig, ax = plt.subplots(figsize=(W,H))
ax.set_xticks([0, W])
ax.set_yticks([0, H])
ax.axhline(Height)
plt.text(4.1,Height,Height,size='10')
ax.grid()
    
for n in range(B):
    Rect_place(Rect2,n)

# 処理後の時刻
t2 = time.time()
 
# 経過時間を表示
elapsed_time = t2-t1
print(f"経過時間：{elapsed_time: .2f}秒")
    
#図形を描画
print(f"高さ：{Height}")
plt.show()
```

```python
#ランダムに並び替えてその中から最も低いパターンを表示する。

# 処理前の時刻
t1 = time.time() 

H_lst = []
j = 100
for i in range(j):
    clear_output(True)
    print(i+1,"/",j)
    Rect4 = Rect.sample(frac=1, ignore_index=True,random_state=i)
    for k in range(B):
        place_coordinate(Rect4,k)
    H_lst.append(Rect4.max()["y+h"])
clear_output()
num = H_lst.index(min(H_lst))

Rect4 = Rect.sample(frac=1, ignore_index=True,random_state=num)

for j in range(B):
    place_coordinate(Rect4,j)

#高さ
Height = round(Rect4.max()["y+h"],2)
H = math.ceil(Height)
#Axesオブジェクトを作成
fig, ax = plt.subplots(figsize=(W,H))
ax.set_xticks([0, W])
ax.set_yticks([0, H])
ax.axhline(Height)
plt.text(4.1,Height,Height,size='10')
ax.grid()

for n in range(B):
    Rect_place(Rect4,n)

# 処理後の時刻
t2 = time.time()
 
# 経過時間を表示
elapsed_time = t2-t1
print(f"経過時間：{elapsed_time: .2f}秒")

#図形を描画
print(f"random_state={num}")
print(f"高さ：{Height}")

plt.show()
```
