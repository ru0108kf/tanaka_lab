# 遺伝的アルゴリズムをDEAPを使って理解する。

## 2.deapのOnemax問題を解読する。
### 設定

```python
# モジュールのインポート
import random

from deap import base
from deap import creator
from deap import tools
import matplotlib.pyplot as plt
%matplotlib inline
```

#### creator.create()関数
遺伝子(設計変数)のセットを表す個体を定義するのに、**creator**というモジュールが用意されており、複雑な構造の個体を表現できる。
例えば、遺伝子のセットを保存したい場合、listを取りあえず使えばいいと思うが、遺伝子が保存されたlistは個体を表しているので、適応度も一緒に保存できない。その場合に、creator.create()関数を使うと、listに適応度(fitness)を保存するメンバ変数を、簡単に追加することができる。

**creator.create**()関数は、あるクラスにメンバ変数を追加して子クラスを新たに作成する関数。\
creator.create()関数を使う時は、少なくとも2個の引数を指定する必要がある。\
第2引数で指定したクラスを継承して、第1引数で指定する名前のクラスをcreatorモジュール内に新しく生成する。\
第3引数以降は、生成する子クラスに追加するメンバ変数を指定する。
  
#### 適応度の定義
1行目の記述は、base.Fitnessクラスを継承して、weights=(1.0,)というメンバ変数を追加した適応度を表すFitnessMaxというクラスをcreatorモジュール内に作成している。

weightsは評価する目的関数が1つの場合で、個体の適応度を最大化したい場合に1.0に設定する。最小化問題ではweightsを-1.0にする。\
基本は2つ以上の引数を扱い、単目的最適化の場合は、引数が一つなので、(-1.0,)とする。,に注意。

#### 個体の定義
2行目の記述は、listクラスを継承して、fitness=creator.FitnessMaxというメンバ変数を追加したIndividualクラスを作成している。

遺伝子を保存する個体listに、適応度fitnessというメンバ変数を追加してIndividualとしている。

```python
creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)
```

### 初期条件
**Toolbox.register**は引数のデフォルト値が無い関数に、デフォルト値を設定することができる。

#### 遺伝子を作成する関数
**attr_bool**をrandom.randintに0,1の引数を与えるという関数にしている。結果、attr_boolは0か1かをランダムで生成する関数となる。
#### 個体を作成する関数
**initRepeat**は、3つの引数container、func、nを持っており、n回funcを実行して、その値をcontainerに格納して返す関数。つまり、100回**toolbox.attr_bool**を実行し、その値を**creator.Individual**に格納して返す関数**individual**を作成する。引数なしで個体を生成できる**individual**関数となる。
#### 世代を作成する関数
個体をtoolbox.individualで作成してlistに格納し、世代を生成するpopulation関数を作っている。世代内の個体数にあたるnは、後でmain関数内で指定するので、デフォルト値は設定していない。

```python
toolbox = base.Toolbox()
# 属性を生成
toolbox.register("attr_bool", random.randint, 0, 1)
# 初期条件を構築
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_bool, 100)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
```

### 評価関数

評価関数は単純で、個体内の1の数。評価関数の返り値に","があるのに注意。おそらく、多目的の場合に返り値が2つ以上になる場合を想定している？

```python
# 評価関数
def evalOneMax(individual):
    return sum(individual),
```

### 遺伝的演算子
評価、交差、突然変異、選択淘汰の関数の登録。\
交叉、突然変異、選択の関数は、DEAPに適当なものを選ぶ。なければ自分で作成する。

```python
# 評価関数、交差、突然変異、選択淘汰
toolbox.register("evaluate", evalOneMax)
toolbox.register("mate", tools.cxTwoPoints)
toolbox.register("mutate", tools.mutFlipBit, indpb=0.05)
toolbox.register("select", tools.selTournament, tournsize=3)
```

#### 交叉関数
二つの個体ind1, ind2を引数として、固体内の遺伝子を2点交叉で入れ替えて返す関数。

#### 突然変異関数
一つの個体individualと遺伝子突然変異率indpbを引数にして、固体内の遺伝子を突然変異させて返す関数。\
遺伝子が0と1のビット値なので、ビット値を入れ替えることで突然変異を表している。\
indpb=0.05というデフォルト値を与えている。遺伝子突然変異率は5%ということ。

#### 選択関数
選択淘汰を実施する関数。引数はindividuals, k, tournsize, fit_attr="fitness"の4つ。\
individualsは、個体individualの後ろにsがついており複数形になっており、この引数には世代(population)を入れる。\
世代の数の分上位からランダムに選択する。\
kは、世代individuals中の個体の数。tournsizeはトーナメントサイズ。
tournsize=3でというデフォルト値を与えている。上から３番目の評価値を持つ個体が生き残るということ。\
fit_attr="fitness"は、固体のメンバ変数になっている適応度の変数名を指定している。fit_attrはfitness_attribute(適応度_属性)の略。\
デフォルト値がfitness。8行目の個体定義時に、fitnessという名前以外で適応度を登録した場合、ここで変えないとエラーになる。

### 実際の進化計算
設定と、評価関数、演算子を選択したら、アルゴリズムを定義する。一般に**main**()という名前を付ける。
- 流れとしては、
- 初期個体群の生成→評価→次世代個体群の生成→交配、変異→評価→次世代の生成
- 条件を満たすまで繰り返す

```python
def main():
    
    # 初期の個体群を生成
    # 初めに個体群をインスタンス化する
    # n=300は１世代の個体の数を表す
    pop = toolbox.population(n=300)
    # CXPBは2つの個体が交配、交差する確率
    # MUTPB は個体が突然変異を起こす確率
    CXPB, MUTPB = 0.5, 0.2

    print("Start of evolution")

    # 初期の個体群の評価
    # すべての個体の数値を評価関数に入れて、リストに格納
    fitnesses = list(map(toolbox.evaluate, pop))
    # 個体と適合をそれぞれ関連付ける
    # fitとpopの順序は同じであることに注意。
    for ind, fit in zip(pop, fitnesses):
        ind.fitness.values = fit

    print("  Evaluated %i individuals" % len(pop))
    
    #　すべての個体の適合度を配列にする
    fits = [ind.fitness.values[0] for ind in pop]
    # 世代数を記録する変数
    g = 0
    log = [[0,min(fits),max(fits)]]

    # 進化計算開始
    while max(fits) < 100 and g < 100:
        # 新しい世代
        g = g + 1
        print("-- Generation %i --" % g)

        # 次世代の個体群を選択（offspringは子孫）
        offspring = toolbox.select(pop, len(pop))
        # 個体群のクローンを生成
        offspring = list(map(toolbox.clone, offspring))

        # 選択した個体群に交差と突然変異を適応する
        # 単純なGAでは、母集団の中で親を直接子孫に置き換える
        # 偶数番目と奇数番目の個体を取り出して交差
        for child1, child2 in zip(offspring[::2], offspring[1::2]):
            # 確率CXPBで2個体を交配させる
            if random.random() < CXPB:
                toolbox.mate(child1, child2)
                # 交配された個体の適合度を削除
                # 後で再計算する必要がある
                del child1.fitness.values
                del child2.fitness.values
        # 確率MUTPBで個体を変異させる
        for mutant in offspring:
            if random.random() < MUTPB:
                toolbox.mutate(mutant)
                # 変異された個体の適合度を削除
                del mutant.fitness.values

        # ここで母集団を再評価する必要があるので、
        # 適合度を削除した個体について適応度の再評価を行う
        invalid_ind = [ind for ind in offspring if not ind.fitness.valid]
        fitnesses = map(toolbox.evaluate, invalid_ind)
        for ind, fit in zip(invalid_ind, fitnesses):
            ind.fitness.values = fit

        print("  Evaluated %i individuals" % len(invalid_ind))

        # そして最後に、古い個体群を子孫に置き換える。
        pop[:] = offspring

        # すべての個体の適合度を配列にする
        fits = [ind.fitness.values[0] for ind in pop]
        
        # 進化は定義された世代数を達成するまで続ける
        
        # 母集団の最小、最大、平均、標準偏差を表示
        length = len(pop)
        mean = sum(fits) / length
        sum2 = sum(x*x for x in fits)
        std = abs(sum2 / length - mean**2)**0.5

        print("  Min %s" % min(fits))
        print("  Max %s" % max(fits))
        print("  Avg %s" % mean)
        print("  Std %s" % std)
        
        log.append([g,min(fits),max(fits)])

    print("-- End of (successful) evolution --")
    
    best_ind = tools.selBest(pop, 1)[0]
    print("Best individual is %s, %s" % (best_ind, best_ind.fitness.values))
    
    return log
```

#### 実装

```python
if __name__ == "__main__":
    log = main()
```

```python
x,y1,y2 = [],[],[]
for i in range(len(log)):
    x.append(log[i][0])
    y1.append(log[i][1])
    y2.append(log[i][2])
    
fig, ax = plt.subplots()
plt.xlim(0,len(x))
plt.ylim(20,100)
line = ax.plot(x, y1, 's-r',x, y2, 'o-b')
plt.title("oneMax") 
ax.set_xlabel("Generation")
ax.set_ylabel("Fitness")

plt.show()
```
