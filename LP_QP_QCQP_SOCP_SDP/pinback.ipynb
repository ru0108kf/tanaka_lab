```python
#モジュールの読み込み
import matplotlib.pyplot as plt
from matplotlib import patches
import math
import numpy as np
import random
import pandas as pd
import itertools
%matplotlib inline
```

```python
#seed値を設定（ここをなくせば完全ランダムになる。色もランダムになる）
seed = 1
random.seed(seed)
np.random.seed(seed)
```

```python
#数値(W:部屋の幅,H:部屋の高さ,B:長方形の数,V:長方形の面積の合計)
W = 4
H = 8
B = 20
V = 20
wh = np.random.uniform(low=0.5, high=1.5, size=(B, 2))
wh = wh*(math.sqrt(V/np.sum(wh[:,0]*wh[:,1])))
xy = [[0,0]]*B
```

```python
#DataFrame処理
wh_df = pd.DataFrame(data=(wh),columns=("w","h"))
xy_df = pd.DataFrame(data=(xy),columns=("x","y"))
Rect = pd.concat([wh_df,xy_df],axis=1)
Rect
```
